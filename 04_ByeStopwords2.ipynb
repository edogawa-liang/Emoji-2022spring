## 選擇停用詞

```python
stopwords = ["的", "了", "之", "你", "我", "他", "她", "它", "祂", "牠", "妳","你們", "我們", "他們", "妳們", 
             "上", "下", "左", "右", "和", "且",  "與", "或", "這", "那", "這裡", "那裡", 
             "上面", "下面", "左邊", "右邊", "前面", "後面", "這個", "那個", ",", "丶"]

print(stopwords)
```

## 去停用詞 + 詞頻統計 + 句長統計 gogo

#### 若第一次執行的是空list，沒辦法蓋過去int(0)
#### 若第一次執行的list有東西，可以蓋過int，且之後的空list也都可以蓋過int

```python
time_start = time.time() # 開始計時
start_ID = 200 
end_ID = 299

for i in range(start_ID, (end_ID + 1)):  
    # 匯入檔案 
    file = "rawno_c" + str(i) + ".xlsx"
    #file = "c" + str(i) + ".xlsx"
    df = pd.read_excel(path_from + file)
    
    df1 = df[['ID', 'content']]
    #df1 = df[['ID', 'content', 'emoji']]
    df1['NoStopCut'] = 0
    df1['NoStopPOS'] = 0
    df1['Length'] = 0
    index= []
    
    cutlist= df['Cut'].tolist()
    poslist= df['POS']
    
    blank= []
    for j in range(len(cutlist)):
        clist = eval(cutlist[j])
        pos = poslist[j].replace("[", "").replace("]", "").replace(",", "").replace("'", "").split(" ")

        bye =  [w for w in clist if w in stopwords] #句子中的停用詞
        
        # if 句子裡有停用詞
        if(len(bye)!= 0):

            lbye = [clist.index(i) for i in bye] #句子中的停用詞的位置

            for k in sorted(range(len(bye)), reverse= True):
                pos.remove(pos[lbye[k]])
                clist.remove(bye[k])
            
            # 若第一筆資料只有停用詞的話會出錯...
            if(clist==[] and j==0):
                blank.append(j)
                continue
            
            else:
                df1['NoStopCut'][j]= clist
                if(clist==[]):
                    blank.append(j)
            
        # 句子內沒有停用詞
        else:
            df1['NoStopCut'][j] = eval(df['Cut'][j])      
        
        df1['NoStopPOS'][j] = pos
        df1['Length'][j]= len(df1['NoStopCut'][j])
        
        index = index + df1['NoStopCut'][j]
    
    length= pd.DataFrame(columns= [i])
    length[i]= df1['Length'].value_counts()
    length.insert(0, 'Length', length.index)
    length= length.sort_values(by= ['Length'])
    
    # 計數
    index = pd.DataFrame(index)
    index = index[0].value_counts()
    index = pd.DataFrame(index)
    index.columns = [i]
    index.insert(0, 'Word', index.index)
    
        
    df1 = df1.drop(blank, axis=0)
    
    df1.to_excel(path_go + f"Raw_nsw_c{i}.xlsx")
    length.to_excel(path_go + f"Raw_nsw_l{i}.xlsx", index= False)
    index.to_excel(path_go + f"Raw_nsw_i{i}.xlsx", index= False)
    
    #df1.to_excel(path_go + f"emoji_nsw_c{i}.xlsx")
    #length.to_excel(path_go + f"emoji_nsw_l{i}.xlsx", index= False)
    #index.to_excel(path_go + f"emoji_nsw_i{i}.xlsx", index= False)
    
    
    time_end = time.time() 
    timespend = (time_end- time_start)/60
    print('檔案 ', i, " 已處理完成. ", "時間: ", round(timespend, 2), " 分鐘")
```
