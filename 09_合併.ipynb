## 看表情符號的數量與分佈

這邊主要是想看表情符號的數量與分布, 所以**斷句**完的就可以了, **斷詞**完的檔案應該很大, 就不麻煩了.

```python
import pandas as pd
```

## 匯入檔案, 並串成一個 data.frame

```python
icon_PATH_in = "C:/Users/e2414/OneDrive/桌面/專題/AllData/emojiclean/" # 匯入路徑
icon_PATH_out = "C:/Users/e2414/OneDrive/桌面/專題/AllData/emojiclean/" # 匯出路徑
```

```python
# 注意: start_ID 到 end_ID 必須連續
start_ID = 200 # 第一個檔案 ID
end_ID = 275 # 最後一個檔案 ID

# 匯入第一個檔案
data = pd.read_excel(icon_PATH_in + str(start_ID) + ".xlsx")

# 匯入後續檔案
for i in range((start_ID + 1), (end_ID + 1)) :
    df = pd.read_excel(icon_PATH_in + str(i) + ".xlsx") # 匯入檔案
    data = data.append(df) # df 與 data 串接

# 因為我的 FB post 中間有無資料的檔案, 所以分段處理
start_ID = 1200 # 第一個檔案 ID
end_ID = 1265 # 最後一個檔案 ID
for i in range(start_ID, (end_ID + 1)) :
    df = pd.read_excel(icon_PATH_in + str(i) + ".xlsx") # 匯入檔案
    data = data.append(df) # df 與 data 串接

start_ID = 280 # 第一個檔案 ID
end_ID = 298 # 最後一個檔案 ID
for i in range(start_ID, (end_ID + 1)) :
    df = pd.read_excel(icon_PATH_in + str(i) + ".xlsx") # 匯入檔案
    data = data.append(df) # df 與 data 串接

# 看一下 data 總數對不對
len(data.index)
```

## 整理 icon

```python
# 先匯入 icon 對照表
all_icon = pd.read_excel(icon_PATH_in + "all_icon.xlsx")

icon = pd.DataFrame(data['emoji'])
icon = pd.DataFrame(icon['emoji'].value_counts())
icon.rename(columns = {'emoji': 'Freq'}, inplace = True) # 更改 colname
icon['emoji'] = icon.index # 把 index 變成變數
last_col = icon.pop(icon.columns[-1]) # 調整 col 順序
icon.insert(0, last_col.name, last_col)
icon['prob'] = "" # icon 的機率
icon['HTML'] = "" # icon 的 HTML 編碼
icon['Group'] = "" # icon 的分組(因為還沒分好組, 所以之後會先是空的)
icon['Label'] = "" # icon 的介紹

for i in range(0, len(icon.index)) :
    icon['HTML'][i] = ord(icon['emoji'][i])
    for s in range(0, len(all_icon.index)) :
        if (icon['HTML'][i] == all_icon['HTML'][s]) :
            icon['Label'][i] = all_icon['Label'][s]
            break

icon = icon.sort_values(by = ['Freq'], ascending= False)
icon['prob'] = icon['Freq'] / sum(icon['Freq'])
icon.to_excel(icon_PATH_out + "icon.xlsx")
```
