## 自動登入

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import time

# 設定網址
url = 'https://www.facebook.com'  

# 開啟 Chrome、前往網址 
driver = webdriver.Chrome()        
driver.get(url)   
time.sleep(0.2)


# 輸入帳號密碼
account= '自行輸入...'
password= '自行輸入...'

## 帳號
elem = driver.find_element_by_id("email")
elem.send_keys(account)

## 密碼
elem = driver.find_element_by_id("pass")
elem.send_keys(password)  
  
# 登入 
elem.send_keys(Keys.RETURN)
time.sleep(5)
```

## 點擊貼文顯示更多、展開回覆
- ### 可改成設定跑的時間或貼文數量( 預設是一直跑下去)
- ### 跑一陣子就先按暫停爬下來，點擊太久網頁會 out of memory 白跑了<br>
如果有按鈕只有xx則回覆，不是檢視xx則回覆，就把回覆加上去ㄅ不然不會點~ 但也有隱藏xx則回覆的按鈕，要小心，加了回覆會一直重複點隱藏回覆檢視回覆隱藏檢視...

```python
import re

#點擊顯示更多、展開回覆
def click_expand_posts():
    try:
        SCROLL_UPPER = -300  # 如果還是有按不到的問題，調整這個參數，決定往上scroll多少pixel
        SCROLL_DOWN = 200    # 如果頁面沒有捕捉到任何元素，往下scroll 200 pixel
        elements = driver.find_elements_by_xpath('//div[@class="j83agx80 buofh1pr jklb3kyz l9j0dhe7"]|//div[@class="oajrlxb2 g5ia77u1 qu0x051f esr5mh6w e9989ue4 r7d6kgcz rq0escxv nhd2j8a9 nc684nl6 p7hjln8o kvgmc6g5 cxmmr5t8 oygrvhab hcukyx3x jb3vyjys rz4wbd8a qt6c0cv9 a8nywdso i1ao9s8h esuyzwwr f1sip0of lzcic4wl gpro0wi8 oo9gr5id lrazzd5p"]|//div[@class="rq0escxv l9j0dhe7 du4w35lb j83agx80 cbu4d94t pfnyh3mw d2edcug0 hpfvmrgz n8tt0mok hyh9befq r8blr3vg jwdofwj8 g0qnabr5"]')
        if elements:
            for ele in elements:
                # 如果有按鈕只有xx則回覆，不是檢視xx則回覆，就把回覆加上去ㄅ不然不會點~ 
                # 但也有隱藏xx則回覆的按鈕，要小心，加了回覆會一直重複點隱藏回覆檢視回覆隱藏檢視...
                if bool(re.search('顯示更多|查看|檢視', ele.text)) == True:
                    print(ele.text)
                    ele.click()
                    time.sleep(2)
        else:
            driver.execute_script(f"window.scrollBy(0, {SCROLL_DOWN})")
    except Exception as e:
        print("errorrrrrrr, retry again...")
        driver.execute_script(f"window.scrollBy(0, {SCROLL_UPPER})")
        click_expand_posts() 

        
#截圖
def screenShot_current_posts():
    cnt = 1
    for ele in driver.find_elements_by_xpath('//div[@class="j83agx80 buofh1pr jklb3kyz l9j0dhe7"]|//div[@class="oajrlxb2 g5ia77u1 qu0x051f esr5mh6w e9989ue4 r7d6kgcz rq0escxv nhd2j8a9 nc684nl6 p7hjln8o kvgmc6g5 cxmmr5t8 oygrvhab hcukyx3x jb3vyjys rz4wbd8a qt6c0cv9 a8nywdso i1ao9s8h esuyzwwr f1sip0of lzcic4wl gpro0wi8 oo9gr5id lrazzd5p"]|//div[@class="rq0escxv l9j0dhe7 du4w35lb j83agx80 cbu4d94t pfnyh3mw d2edcug0 hpfvmrgz n8tt0mok hyh9befq r8blr3vg jwdofwj8 g0qnabr5"]'):
        ele.screenshot(f"screenshot_{cnt}.png")
        cnt += 1
        
import time
timespend= 1    
time_start = time.time() #開始計時


#while timespend< 22:  # 我的電腦有顧跑約22分鐘爆掉...
# 可以改成設定數字就不會一直跑下去，但網頁碼比較奇怪沒辦法設定確切要點擊幾篇(或我不會)...
# try: 設定4篇只會點1篇、設8篇點4篇、設15篇點13篇，建議設定n+5
#while len(driver.find_elements_by_xpath('//div[@class="j83agx80 cbu4d94t"]'))< 10+5:
while True:
    try:
        click_expand_posts()
        print('**偵測到{}篇貼文**'.format(len(driver.find_elements_by_xpath('//div[@class="j83agx80 cbu4d94t"]'))))
        time_end = time.time() 
        timespend= (time_end- time_start)/60
        print('spend '+ str(round(timespend, 2))+ ' min')
    
    #當按暫停鍵時
    except KeyboardInterrupt:
        print("force to stop by user")
        break
    
    #當還是有按不到的問題時，截圖結束()
    except Exception as e:
        print(e)
        screenShot_current_posts()
        break
```

## 抓出整個貼文+留言框框

```python
htmltext = driver.page_source
soup = BeautifulSoup(htmltext,"lxml")
postL = soup.find_all('div','j83agx80 cbu4d94t')

# 總共讀到幾篇文
postList= postL[:len(postL)]
print(len(postList))
```

## 印出貼文 (只抓本人po的文章)

```python
import pandas as pd

pattern= r"(?<=img alt=\")\S|[\u4e00-\u9fa5]+"

df1 = pd.DataFrame(columns=['siteName', 'postID',#'postName', 
                            'time', 'post'])

postId= 0
for post in postList:
    name_check= post.find('a', 'oajrlxb2 g5ia77u1 qu0x051f esr5mh6w e9989ue4 r7d6kgcz rq0escxv nhd2j8a9 nc684nl6 p7hjln8o kvgmc6g5 cxmmr5t8 oygrvhab hcukyx3x jb3vyjys rz4wbd8a qt6c0cv9 a8nywdso i1ao9s8h esuyzwwr f1sip0of lzcic4wl gpro0wi8 oo9gr5id lrazzd5p').text
    
    # 只抓是本人po的文章
    if name_check== siteName: 
        name= post.find('a', 'oajrlxb2 g5ia77u1 qu0x051f esr5mh6w e9989ue4 r7d6kgcz rq0escxv nhd2j8a9 nc684nl6 p7hjln8o kvgmc6g5 cxmmr5t8 oygrvhab hcukyx3x jb3vyjys rz4wbd8a qt6c0cv9 a8nywdso i1ao9s8h esuyzwwr f1sip0of lzcic4wl gpro0wi8 oo9gr5id lrazzd5p').text
        time = post.find('span','tojvnm2t a6sixzi8 abs2jz4q a8s20v7p t1p8iaqh k5wvi7nf q3lfd5jv pk4s997a bipmatt0 cebpdrjk qowsmv63 owwhemhu dp1hu0rb dhp61c6y iyyx5f41').text
        time= re.sub('=', '', time)

        # content可能需要改網頁碼
        content = post.find('span','d2edcug0 hpfvmrgz qv66sw1b c1et5uql lr9zc1uh a8c37x1j fe6kdd0r mau55g9w c8b282yb keod5gw0 nxhoafnm aigsh9s9 d3f4x2em iv3no6db jq4qci2q a3bd9o3v b1v8xokw oo9gr5id hzawbc8m')

        if content is not None:
            content = ','.join(re.findall(pattern, str(content)))
            
        postId= postId+ 1 

        df1 = df1.append({
        'siteName': siteName, 
        'postID':postId, 
        #'postName': name,
        'time' :time ,
        'post' : content} , ignore_index= True)
```

## 印出回覆 (不爬replyName)

```python
#印出回覆
df2 = pd.DataFrame(columns=['siteName', 'postID',#'replyName', 
                            'replyOrder', 'reply'])

namelen= len(siteName)
postID= 1
for reply in postList:
    name_check= reply.find('h2', 'gmql0nx0 l94mrbxd p1ri9a11 lzcic4wl aahdfvyu hzawbc8m').text
    if name_check[0:namelen] == siteName:
        userReplys = reply.find_all('div','tw6a2znq sj5x9vvc d1544ag0 cxgpxx05')
        
        order= 1
        for userReply in userReplys:
            replyName= userReply.find('span','d2edcug0 hpfvmrgz qv66sw1b c1et5uql lr9zc1uh a8c37x1j fe6kdd0r mau55g9w c8b282yb keod5gw0 nxhoafnm aigsh9s9 d9wwppkn mdeji52x e9vueds3 j5wam9gi lrazzd5p oo9gr5id').text
            
            # content可能需要改網頁碼
            content= userReply.find_all('div',{'dir':'auto', 'style':'text-align: start;'})

            if content is not None:
                content = ','.join(re.findall(pattern, str(content)))
            
            df2 = df2.append({
                'siteName': siteName,
                'postID': postID, 
                #'replyName' : replyName,
                'replyOrder': order,
                'reply' : content} , ignore_index= True)
            order= order+ 1

        postID= postID+ 1
```
