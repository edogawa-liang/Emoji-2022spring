## æŠ½æ¨£ fun

```python
def sampling(train, trainlabel, group, sampling):
    
    alldata = list(range(np.shape(train)[0]))  #è³‡æ–™ç¸½å…±å¹¾ç­†
   
    # è¡¨æƒ… group å°æ‡‰çš„ä½ç½®
    allemo = []
    for j in range(1, group + 1):
        j_emo = [i for i in alldata if trainlabel[i]== j]
        allemo.append(j_emo)
        # allemo:[[group 1 è¡¨æƒ…ç¬¦è™Ÿä½ç½®], [group 2 è¡¨æƒ…ç¬¦è™Ÿä½ç½®], [], ...[group 40 è¡¨æƒ…ç¬¦è™Ÿä½ç½®]]
    
    
    #æŠ½æ¨£
    random_want = []
    for j in range(group):
        gg = np.random.choice(allemo[j], sampling, replace = False).tolist() #æ¯å€‹çµ„éš¨æ©ŸæŠ½åˆ°çš„è¡¨æƒ…ä½ç½®
        random_want = random_want + gg
        # random_want: [group 1 æŠ½æ¨£çš„ä½ç½®, group 2 æŠ½æ¨£çš„ä½ç½® ...]
        # åŒ¯å‡ºæŒ‰ç…§ groupæ’åºï¼Œlabelå¯ç›´æ¥ä¸Šç‚º 111..11222..22333...33 .....
        
    return allemo, random_want
```

```python
sentence = testdata['å¥']
emo = testdata['emoji']
group = [int(i) for i in testdata['Group'].tolist()]
oo = testdata['å¥å­ä½ç½®']
```

```python
%%time
result = sampling(sentence, group, 40, 2)
```

```python
# è™•ç†å¥å­
sen = []
for i in range(len(sentence[result[1]])):
        sen.append(''.join(eval(sentence[result[1]].tolist()[i])))
        
```
