## 抽樣 fun

```python
def sampling(train, trainlabel, group, sampling):
    
    alldata = list(range(np.shape(train)[0]))  #資料總共幾筆
   
    # 表情 group 對應的位置
    allemo = []
    for j in range(1, group + 1):
        j_emo = [i for i in alldata if trainlabel[i]== j]
        allemo.append(j_emo)
        # allemo:[[group 1 表情符號位置], [group 2 表情符號位置], [], ...[group 40 表情符號位置]]
    
    
    #抽樣
    random_want = []
    for j in range(group):
        gg = np.random.choice(allemo[j], sampling, replace = False).tolist() #每個組隨機抽到的表情位置
        random_want = random_want + gg
        # random_want: [group 1 抽樣的位置, group 2 抽樣的位置 ...]
        # 匯出按照 group排序，label可直接上為 111..11222..22333...33 .....
        
    return allemo, random_want
```

```python
sentence = testdata['句']
emo = testdata['emoji']
group = [int(i) for i in testdata['Group'].tolist()]
oo = testdata['句子位置']
```

```python
%%time
result = sampling(sentence, group, 40, 2)
```

```python
# 處理句子
sen = []
for i in range(len(sentence[result[1]])):
        sen.append(''.join(eval(sentence[result[1]].tolist()[i])))
        
```

```python
#dic
survey = pd.DataFrame({'original 中的句子位置': [oo[i] for i in result[1]],
                      'test 中的句子位置':result[1], 
                      '句子': sen, 
                      'Group': [group[i] for i in result[1]],
                      'emoji': [emo[i] for i in result[1]]})
```
