```python
import re

#é»æ“Šé¡¯ç¤ºæ›´å¤šã€å±•é–‹å›è¦†
def click_expand_posts():
    try:
        SCROLL_UPPER = -300  # å¦‚æœé‚„æ˜¯æœ‰æŒ‰ä¸åˆ°çš„å•é¡Œï¼Œèª¿æ•´é€™å€‹åƒæ•¸ï¼Œæ±ºå®šå¾€ä¸Šscrollå¤šå°‘pixel
        SCROLL_DOWN = 200    # å¦‚æœé é¢æ²’æœ‰æ•æ‰åˆ°ä»»ä½•å…ƒç´ ï¼Œå¾€ä¸‹scroll 200 pixel
        elements = driver.find_elements_by_xpath('//div[@class="j83agx80 buofh1pr jklb3kyz l9j0dhe7"]|//div[@class="oajrlxb2 g5ia77u1 qu0x051f esr5mh6w e9989ue4 r7d6kgcz rq0escxv nhd2j8a9 nc684nl6 p7hjln8o kvgmc6g5 cxmmr5t8 oygrvhab hcukyx3x jb3vyjys rz4wbd8a qt6c0cv9 a8nywdso i1ao9s8h esuyzwwr f1sip0of lzcic4wl gpro0wi8 oo9gr5id lrazzd5p"]|//div[@class="rq0escxv l9j0dhe7 du4w35lb j83agx80 cbu4d94t pfnyh3mw d2edcug0 hpfvmrgz n8tt0mok hyh9befq r8blr3vg jwdofwj8 g0qnabr5"]')
        if elements:
            for ele in elements:
                #æœ‰æ™‚å€™æœƒå‡ºç¾éš±è—xxå‰‡å›è¦† å°±æœƒä¸€ç›´é‡è¤‡é»éš±è—æª¢è¦–éš±è—æª¢è¦–...é‚£æˆ‘ä¹Ÿæ²’è¾¦æ³•...
                #å…ˆæŒ‰æš«åœè®“ä»–æ»‘éå†ç¹¼çºŒã„… 
                if bool(re.search('é¡¯ç¤ºæ›´å¤š|æŸ¥çœ‹|æª¢è¦–|å›è¦†', ele.text)) == True:
                    print(ele.text)
                    ele.click()
                    time.sleep(2)
        else:
            driver.execute_script(f"window.scrollBy(0, {SCROLL_DOWN})")
    except Exception as e:
        print("errorrrrrrr, retry again...")
        driver.execute_script(f"window.scrollBy(0, {SCROLL_UPPER})")
        click_expand_posts() 

        
#æˆªåœ–
def screenShot_current_posts():
    cnt = 1
    for ele in driver.find_elements_by_xpath('//div[@class="j83agx80 buofh1pr jklb3kyz l9j0dhe7"]|//div[@class="oajrlxb2 g5ia77u1 qu0x051f esr5mh6w e9989ue4 r7d6kgcz rq0escxv nhd2j8a9 nc684nl6 p7hjln8o kvgmc6g5 cxmmr5t8 oygrvhab hcukyx3x jb3vyjys rz4wbd8a qt6c0cv9 a8nywdso i1ao9s8h esuyzwwr f1sip0of lzcic4wl gpro0wi8 oo9gr5id lrazzd5p"]|//div[@class="rq0escxv l9j0dhe7 du4w35lb j83agx80 cbu4d94t pfnyh3mw d2edcug0 hpfvmrgz n8tt0mok hyh9befq r8blr3vg jwdofwj8 g0qnabr5"]'):
        ele.screenshot(f"screenshot_{cnt}.png")
        cnt += 1
        
import time
timespend= 1    
time_start = time.time() #é–‹å§‹è¨ˆæ™‚


#while timespend< 22:  # æˆ‘çš„é›»è…¦æœ‰é¡§è·‘ç´„22åˆ†é˜çˆ†æ‰...
# å¯ä»¥æ”¹æˆè¨­å®šæ•¸å­—å°±ä¸æœƒä¸€ç›´è·‘ä¸‹å»ï¼Œä½†ç¶²é ç¢¼æ¯”è¼ƒå¥‡æ€ªæ²’è¾¦æ³•è¨­å®šç¢ºåˆ‡è¦é»æ“Šå¹¾ç¯‡(æˆ–æˆ‘ä¸æœƒ)...
# try: è¨­å®š4ç¯‡åªæœƒé»1ç¯‡ã€è¨­8ç¯‡é»4ç¯‡ã€è¨­15ç¯‡é»13ç¯‡ï¼Œå»ºè­°è¨­å®šn+5
#while len(driver.find_elements_by_xpath('//div[@class="j83agx80 cbu4d94t"]'))< 10+5:
while True:
    try:
        click_expand_posts()
        print('**åµæ¸¬åˆ°{}ç¯‡è²¼æ–‡**'.format(len(driver.find_elements_by_xpath('//div[@class="j83agx80 cbu4d94t"]'))))
        time_end = time.time() 
        timespend= (time_end- time_start)/60
        print('spend '+ str(round(timespend, 2))+ ' min')
    
    #ç•¶æŒ‰æš«åœéµæ™‚
    except KeyboardInterrupt:
        print("force to stop by user")
        break
    
    #ç•¶é‚„æ˜¯æœ‰æŒ‰ä¸åˆ°çš„å•é¡Œæ™‚ï¼Œæˆªåœ–çµæŸ()
    except Exception as e:
        print(e)
        screenShot_current_posts()
        break
```

```python
import pandas as pd

pattern= r"(?<=img alt=\")\S|[\u4e00-\u9fa5]+"

df1 = pd.DataFrame(columns=['siteName', 'postID',#'postName', 
                            'time', 'post'])

postId= 0
for post in postList:
    name= post.find('a', 'oajrlxb2 g5ia77u1 qu0x051f esr5mh6w e9989ue4 r7d6kgcz rq0escxv nhd2j8a9 nc684nl6 p7hjln8o kvgmc6g5 cxmmr5t8 oygrvhab hcukyx3x jb3vyjys rz4wbd8a qt6c0cv9 a8nywdso i1ao9s8h esuyzwwr f1sip0of lzcic4wl gpro0wi8 oo9gr5id lrazzd5p').text
    time = post.find('span','tojvnm2t a6sixzi8 abs2jz4q a8s20v7p t1p8iaqh k5wvi7nf q3lfd5jv pk4s997a bipmatt0 cebpdrjk qowsmv63 owwhemhu dp1hu0rb dhp61c6y iyyx5f41').text
    time= re.sub('=', '', time)
   
    # contentçš„ç¶²é ç¢¼å¯èƒ½è¦æ”¹
    content = post.find('span','d2edcug0 hpfvmrgz qv66sw1b c1et5uql lr9zc1uh a8c37x1j fe6kdd0r mau55g9w c8b282yb keod5gw0 nxhoafnm aigsh9s9 d3f4x2em iv3no6db jq4qci2q a3bd9o3v b1v8xokw oo9gr5id hzawbc8m')    
    if content is not None:
        content = ','.join(re.findall(pattern, str(content)))
        
    postId= postId+ 1
    
    df1 = df1.append({
    'siteName': siteName, 
    'postID':postId, 
    #'postName': name,
    'time' :time ,
    'post' : content} , ignore_index= True)
```
